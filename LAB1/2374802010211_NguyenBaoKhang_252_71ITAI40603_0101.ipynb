```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('cat.jpg')
img = img[:, :, ::-1]
plt.imshow(img)
```

**1. S dng boxFilter: cv2.blur() hoc cv2.boxFilter()  lm m nh (s dng nhiu thng s filter khc nhau), biu din nh gc v nh lm m cng nhau  kim chng.** 

```python
blur_1 = cv2.blur(img, (3, 3))
blur_2 = cv2.blur(img, (7, 7))
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("nh gc")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(blur_1)
plt.title("Box Filter (3x3)")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(blur_2)
plt.title("Box Filter (7x7)")
plt.axis('off')

plt.show()
```

**2. S dng Gaussian Filter # Lu : kch thc b lc l s l.**

```python
gaussian_1 = cv2.GaussianBlur(img, (5,5), 0)
gaussian_2 = cv2.GaussianBlur(img, (11,11), 0)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("nh gc")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gaussian_1)
plt.title("Gaussian (5x5)")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(gaussian_2)
plt.title("Gaussian (11x11)")
plt.axis('off')

plt.show()
```

**3. S dng Median Filter**

```python
median_1 = cv2.medianBlur(img, 5)
median_2 = cv2.medianBlur(img, 9)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("nh gc")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(median_1)
plt.title("Median Filter (5)")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(median_2)
plt.title("Median Filter (9)")
plt.axis('off')

plt.show()
```

**Given two images as follow (with different sizes: 193x157 and 194x160)**

**a. Crop two images to size 190x155**

```python
img1 = cv2.imread('exT3_01.jpg')
img2 = cv2.imread('exT3_02.jpg')
img1 = img1[:155,1:191]
img2 = img2[:155,:190]

plt.subplot(2,2,1)
plt.imshow(img1)
plt.subplot(2,2,2)
plt.imshow(img2)
```

**b. Transform images to negative ones.**

```python
inv_img1 = cv2.bitwise_not(img1)
inv_img2 = cv2.bitwise_not(img2)
plt.subplot(2,2,3)
plt.imshow(inv_img1)
plt.subplot(2,2,4)
plt.imshow(inv_img2)
```

**c. Process to have an image which has only the ball.**

```python
import cv2
import matplotlib.pyplot as plt

# Read & crop
img1 = cv2.imread('exT3_01.jpg')
img1 = img1[:155, 1:191]

# Convert to grayscale
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Threshold: ball is darker than background
_, mask = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

# Extract ball
ball = cv2.bitwise_and(img1, img1, mask=mask)

# Convert for display
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
ball = cv2.cvtColor(ball, cv2.COLOR_BGR2RGB)

# Show
plt.subplot(1,2,1)
plt.imshow(img1)
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(ball)
plt.title("Only the ball")
plt.axis('off')

plt.show()
```
